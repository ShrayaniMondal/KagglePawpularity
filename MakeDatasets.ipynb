```python
import pandas as pd
#from sklearn import model_selection
import os
from fast_ml.model_development import train_valid_test_split
import os.path
import shutil
```

```python
df = pd.read_csv('train.csv')
df = df.dropna().reset_index(drop=True)

X_train, y_train, X_valid, y_valid, X_test, y_test = train_valid_test_split(df, target = 'Pawpularity', 
                                                                            train_size=0.7, valid_size=0.2, test_size=0.1)
print("Train shape")
print(X_train.shape), print(y_train.shape)
print("Valid shape")
print(X_valid.shape), print(y_valid.shape)
print("Test shape")
print(X_test.shape), print(y_test.shape)
print("OK")
```

```python
dataPath = ['Data/Train', 'Data/Valid','Data/Test']

# Check whether the specified path exists or not
for path in dataPath:
  isExist = os.path.exists(path)
  print(isExist)

  if not isExist:
    os.makedirs(path)
    print("The new directory {} is created!".format(path))
```

```python
src = os.getcwd() + "/train/"
dst = os.getcwd() + "/Data/Train/"
for index, row in X_train.iterrows():
  srcpath = os.path.join(src, row['Id']) + ".jpg"
  dstpath = os.path.join(dst, row['Id']) + ".jpg"
  shutil.copyfile(srcpath, dstpath)
```

```python
src = os.getcwd() + "/train/"
dst = os.getcwd() + "/Data/Valid/"
for index, row in X_valid.iterrows():
  srcpath = os.path.join(src, row['Id']) + ".jpg"
  dstpath = os.path.join(dst, row['Id']) + ".jpg"
  shutil.copyfile(srcpath, dstpath)
```

```python
src = os.getcwd() + "/train/"
dst = os.getcwd() + "/Data/Test/"
for index, row in X_test.iterrows():
  srcpath = os.path.join(src, row['Id']) + ".jpg"
  dstpath = os.path.join(dst, row['Id']) + ".jpg"
  shutil.copyfile(srcpath, dstpath)
```

```python
train_df = X_train
train_df.loc[:, "Pawpularity"] = y_train
print(train_df.shape)

valid_df = X_valid
valid_df.loc[:, "Pawpularity"] = y_valid
print(valid_df.shape)

test_df = X_test
test_df.loc[:, "Pawpularity"] = y_test
print(test_df.shape)
```

```python
metadataPath = 'Data/MetaData'

# Check whether the specified path exists or not
isExist = os.path.exists(metadataPath)
print(isExist)

if not isExist:
  os.makedirs(metadataPath)
  print("The new directory {} is created!".format(metadataPath))
```

```python
train_df.to_csv("Data/MetaData/train.csv", index=False)
valid_df.to_csv("Data/MetaData/valid.csv", index=False)
test_df.to_csv("Data/MetaData/test.csv", index=False)
```
